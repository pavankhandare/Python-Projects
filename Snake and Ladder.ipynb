```python
import random as r
class SnakeLadder:
    player_1_name = ""
    player_2_name = ""
    player_1_pos = 0
    player_2_pos = 0
    curr_player = ""
    roll_num = 0
    snakes = {27:5, 40:3, 43:18, 54:31, 66:45, 76:58, 89:53, 99:41}
    ladders =  {4:25, 13:46, 33:49, 50:69, 62:81, 74:92}
    def __init__(self):
        print("Welcome to Snake and Ladder Game")
        self.player_1_name = input("Enter name of first player :")
        self.player_2_name = input("Enter name of second player :")
        self.roll_dice()
        
    def roll_dice(self):
        if self.player_1_pos>=100:
            print("Hey {} you are winner".format(self.player_1_name))
            return
        elif self.player_2_pos>=100:
            print("Hey {} you are winner".format(self.player_2_name))
            return 
        else:
            if self.curr_player == self.player_1_name:
                dice = input("{} enter 0 to roll the dice".format(self.player_1_name))
                self.roll_num = r.choice([1,2,3,4,5,6])
                if dice:
                    self.player_1_pos += self.roll_num
                    if self.player_1_pos in self.snakes.keys():
                        self.player_1_pos = self.snakes[self.player_1_pos]
                        print("Ohh...there is a snake..you fall down to {}".format(self.player_1_pos))
                        self.curr_player = self.player_2_name
                        self.roll_dice()
                    elif self.player_1_pos in self.ladders.keys():
                        self.player_1_pos = self.ladders[self.player_1_pos]
                        print("Great...you have a ladder..you step up to {}".format(self.player_1_pos))
                        self.curr_player = self.player_2_name
                        self.roll_dice()
                    else:
                        self.curr_player = self.player_2_name
                        self.roll_dice()
                else:
                    print("Wrong input please enter correct input")
                    self.roll_dice()
            else:
                dice = input("{} enter 0 to roll the dice".format(self.player_2_name))
                self.roll_num = r.choice([1,2,3,4,5,6])
                if dice:
                    self.player_2_pos += self.roll_num
                    if self.player_2_pos in self.snakes.keys():
                        self.player_2_pos = self.snakes[self.player_2_pos]
                        print("Ohh...there is a snake..you fall down to {}".format(self.player_2_pos))
                        self.curr_player = self.player_1_name
                        self.roll_dice()
                    elif self.player_2_pos in self.ladders.keys():
                        self.player_2_pos = self.ladders[self.player_2_pos]
                        print("Great...you have a ladder..you step up to {}".format(self.player_2_pos))
                        self.curr_player = self.player_1_name
                        self.roll_dice()
                    else:
                        self.curr_player = self.player_1_name
                        self.roll_dice()
                else:
                    print("Wrong input please enter correct input")
                    self.roll_dice()
```
